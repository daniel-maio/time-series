```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
```

```python
csv_path = Path("../datasets/predictions/deploy").resolve()

print(csv_path)
```

```python
def get_csv(path: Path):

    if not path.exists():
        print("Diretório não existe.")
        return None

    file_path = next(path.glob("*.csv"), None)
    
    if file_path == None:
        print(f"Arquivo não encontrado.")
        return None
    
    else:
        df = pd.read_csv(file_path, parse_dates=True, index_col=0)
        return df
```

```python
df = get_csv(csv_path)
print(df.head())
```

```python
df.index
```

```python
plt.figure()
plt.plot(df['prediction'], marker = 'o', color='b')
plt.title("Sales Predictions")
plt.xlabel("Date")
plt.ylabel("Sales Prediction")
plt.grid("True")
plt.show()
```
