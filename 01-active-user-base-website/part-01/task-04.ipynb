# Suavization

## Simple Moving Average

```python
data['active_users_sma'] = data.active_users.rolling(window=12).mean()
```

```python
data.tail()
```

```python
plt.plot(data['active_users'])
plt.plot(data['active_users_sma'])
plt.show()
```

```python
ts_sma = data['active_users'] - data['active_users_sma']
```

```python
ts_sma.dropna(inplace=True)
```

```python
plt.plot(ts_sma)
plt.show()
```

```python
test_stationarity(ts_sma)
```

## EWMA

```python
data['active_users_ewma'] = data.active_users.ewm(alpha=0.15, adjust=True).mean()
```

```python
ts_ewma = data['active_users'] - data['active_users_ewma']
```

```python
plt.plot(data['active_users'])
plt.plot(data['active_users_ewma'])
plt.show()
```

```python
test_stationarity(ts_ewma)
```
