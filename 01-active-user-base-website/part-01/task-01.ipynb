# Data Processing

```python
data.info()
```

```python
# Convert month column type from object to datetime
data['month'] = pd.to_datetime(data['month'], format="%Y-%m")
```

```python
data.head()
```

```python
# Set the month column as the Index (row label)

data_ts = data.set_index('month')
```

## Boxplot

```python
data_ts.active_users.index.year
```

```python
n_colors = len(np.unique(ar=data_ts.active_users.index.year))
```

```python
colors = sns.color_palette("hls", n_colors)
colors
```

```python
year_index = data_ts.active_users.index.year

sns.boxplot(data=data_ts,
            x = year_index,
            y="active_users",
            hue=year_index,
            palette=colors,
            legend=False)

plt.xlabel("Year")
plt.ylabel("Active Users")
plt.show()
```

# Time Series Decomposition

```python
multiplicative_decomp = sm.tsa.seasonal_decompose(x=data_ts,
                                                  model="multiplicative",
                                                  extrapolate_trend='freq')
```

```python
multiplicative_decomp.plot().suptitle("Multiplicative Decomposition")
plt.show()
```

```python
additive_decomp = sm.tsa.seasonal_decompose(x=data_ts,
                                                  model="additive",
                                                  extrapolate_trend='freq')
```

```python
additive_decomp.plot().suptitle("Additive Decomposition")
plt.show()
```

# Extracting the Components

```python
ts_decomp = pd.concat(
    objs=[
        multiplicative_decomp.trend,
        multiplicative_decomp.seasonal,
        multiplicative_decomp.resid,
        multiplicative_decomp.observed
    ],
    axis=1
)
```

```python
ts_decomp.columns = ["trend", "seasonal", "residuals", "observed"]
```

```python
ts_decomp.head()
```

```python
ts_decomp.to_csv("data/processed/ts_decomposition.csv")
```
