# Transformations

## Log Transformation

```python
active_users_log = np.log(time_series.active_users)
```

```python
plt.plot(active_users_log)
plt.show()
```

```python
test_stationarity(active_users_log)
```

## Square Root Transformation

```python
active_users_sqr = np.sqrt(time_series.active_users)
```

```python
plt.plot(active_users_sqr)
plt.show()
```

```python
test_stationarity(active_users_sqr)
```

## Box-Cox

```python
# lambda = None -> returns maxlog
_, max_log = boxcox(x=time_series['active_users'])
```

```python
max_log
```

```python
active_users_box = boxcox(x=time_series['active_users'], lmbda=max_log)
```

```python
test_stationarity(active_users_box)
```
