```python
import numpy as np
import pandas as pd
```

'day_of_week': Day of the week (0=Sunday, 6=Saturday)

'holiday': # Holiday or not

'active_marketing_campaign': Active marketing campaign or not

'average_conversion_rate': Average conversion rate in %

'number_of_visits_previous_day': # Visits on the previous day

'number_of_visits': np.random.poisson(120, size = data_size) # Number of visits (dependent variable)

```python
rng = np.random.default_rng()
data_size = 1000

df = pd.DataFrame(
    data={
        'day_of_week': rng.integers(low=0, high=6, endpoint=True, size=data_size),
        'holiday': rng.integers(low=0, high=1, endpoint=True, size=data_size),
        'active_marketing_campaign': rng.integers(low=0, high=1, endpoint=True, size=data_size),
        'average_conversion_rate': rng.uniform(low=0, high=50, size=data_size),
        'num_visits_previous_day': rng.poisson(lam=100, size=data_size),
        'num_visits': rng.poisson(lam=120, size=data_size)
    }
)
```

```python
df.to_csv("data/dataset.csv", index=False, header=True)
```
